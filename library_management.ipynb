```python
class Library:
    def __init__(self, data):
        self.booklist = data

    def showBooks(self):
        print("List of books:")
        for book in self.booklist:
            print(book)

    def lendBook(self, bookName):
        if bookName in self.booklist:
            self.booklist.remove(bookName)
            print("Thank you, your request is fulfilled.")
        else:
            print("Book is not available.")

    def returnBook(self, bookName, returnedBookName):
      print(bookName)
      print(returnedBookName)
      if bookName == returnedBookName:
            if returnedBookName not in self.booklist:
                self.booklist.append(returnedBookName)
                print("Thank you for returning the book.")
            else:
                print("Book names do not match.")
      else:
            print("Book names do not match.")

class Customer:
    def reqBook(self):
        book = input("Enter the book name: ")
        return book

lib_Blr = Library(["king lear", "hamlet", "the great gatsby", "ulysses", "all tomorrow parties", "slow man", "the big sheep", "never let me go", "run", "the stand"])
customer_anto = Customer()

while True:
   
    print("1. Show list of books")
    print("2. Lend a book")
    print("3. Return a book")
    print("4. Exit")
    option = input("Enter option number: ")

    if option == "1":
        lib_Blr.showBooks()

    elif option == "2":
        requestedBook = customer_anto.reqBook()
        lib_Blr.lendBook(requestedBook)

    elif option == "3":
        returnedBook = customer_anto.reqBook()
        intendedBook = input("Enter the name of the book that you received: ")
        lib_Blr.returnBook(intendedBook, returnedBook)

    elif option == "4":
        print("Thank you for visiting. Goodbye!")
        break

    else:
        print("Invalid option. Please try again.")
```
